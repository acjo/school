```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.optimize import newton
```

```python
def U_cf(r):
  return 1/(2*(r**2))

def U(r):
  return r**2/2

def U_eff(r):
  return U_cf(r) + U(r)


domain = np.linspace(0.25, 4, 2000)

fig = plt.figure()
fig.set_dpi(150)
ax = fig.add_subplot(111)

ax.plot(domain, U(domain), 'r-', label = r'$U(r)$')
ax.plot(domain, U_cf(domain), 'b-', label = r'$U_{cf}(r)$')
ax.plot(domain, U_eff(domain), 'k-', label=r'$U_{eff}(r)$')
ax.legend(loc='best')
ax.set_title(r'Potential Energies')
ax.set_xlabel('r')
ax.set_ylabel('Potential Energy')
ax.set_xticklabels([])
ax.set_yticklabels([])
plt.plot()

```
