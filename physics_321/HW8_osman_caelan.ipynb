```python
import numpy as np
from matplotlib import pyplot as plt

mg = 1
k = 1

gravity = lambda y: mg*y
spring = lambda y: -0.5*y**2
total = lambda y: gravity(y) + spring(y)

fig = plt.figure()
fig.set_dpi(600)

y_vals = np.linspace(-2, 3, 300)

plt.plot(y_vals, gravity(y_vals), 'r-', lw=1, label ='gravity potential')
plt.plot(y_vals, spring(y_vals), 'b-', lw=1, label='spring potential')
plt.plot(y_vals, total(y_vals), 'm-', lw=1, label='total potential')
plt.xlabel('y(m)')
plt.ylabel('U(J)')
plt.legend(loc='best')
plt.show()
```
